```python
import os
import numpy as np
from PIL import Image
import torch
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
data_path = "./dataset_all/sample_images_from_dataset/"

image_names = ["Cr_10.bmp", "Cr_25.bmp",
              "In_5.bmp", "In_34.bmp",
              "Pa_7.bmp", "Pa_8.bmp",
              "PS_50.bmp", "PS_71.bmp",
              "RS_7.bmp", "RS_8.bmp",
              "Sc_22.bmp", "Sc_23.bmp"]
```

```python
listofimages = [os.path.join(data_path, image_name) for image_name in image_names]
listofimages
```

```python
trans = transforms.ToTensor()
```

```python
imgs = [trans(Image.open(img)) for img in listofimages]
imgs_1 = imgs[0::2]
imgs_2 = imgs[1::2]
imglist = imgs_1 + imgs_2
```

```python
%matplotlib inline
def show(img):
    npimg = img.numpy()
    npimg = np.transpose(npimg, (1,2,0))
    plt.imshow(npimg)
    plt.xticks([])
    plt.yticks([])
    plt.savefig("assets/Collage.eps", quality=100, dpi=400, pad_inches=0, bbox_inches='tight')
```

```python
show(make_grid(imglist, nrow=6, padding=5))
```
